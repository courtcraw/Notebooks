```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# from astropy.timeseries import LombScargle
```

```python
def generate_pysyd_files(star_name,generate_lc_file=False):
    print('Generating',star_name,'pysyd file')
#     if generate_lc_file:
    lightcurve = pd.read_csv('../data/hpfiltered_LCs/KIC'+star_name[4:]+'_LC.csv')
    powerspec = pd.read_csv('../data/hpfiltered_LCs/KIC'+star_name[4:]+'_PS.csv')
    
    eff_obs_time = len(lightcurve) * 1800e-6 #this is in megaseconds, which will inverse to muHz
    print(eff_obs_time)
    
#     print(powerspec['power'])
    powerspec['power'] = powerspec['power']*eff_obs_time #convert to power DENSITY
#     print(powerspec['power'])
    
    if generate_lc_file:
        lightcurve.to_csv('../pysyd-work/data/'+star_name[4:]+'_LC.txt',
                          index=False,header=False, sep ='\t')
#     powerspec[['freq','power']].query('freq < 267').to_csv('../numax_refining/KIC'+star_name[4:]+'_PS.txt',
#                                        index=False,header=False, sep ='\t')
#     powerspec[['freq','power']].query('freq < 267').iloc[::10,:].to_csv('../../pySYD-master/work/data/KIC'+star_name[4:]+'_PS.txt',
#                                        index=False,header=False, sep ='\t')
    powerspec[['freq','power']].query('freq < 267').iloc[::10,:].to_csv('../pysyd-work/data/'+star_name[4:]+'_PS.txt',
                                       index=False,header=False, sep ='\t')
```

```python
def generate_emcee_files(star_name,generate_lc_file=False):
    print('Generating',star_name,'emcee file')
#     if generate_lc_file:
    lightcurve = pd.read_csv('../data/hpfiltered_LCs/KIC'+star_name[4:]+'_LC.csv')
    powerspec = pd.read_csv('../data/hpfiltered_LCs/KIC'+star_name[4:]+'_PS.csv')
    
    eff_obs_time = len(lightcurve) * 1800e-6 #this is in megaseconds, which will inverse to muHz
    print(eff_obs_time)
    
    powerspec['power'] = powerspec['power']*eff_obs_time #convert to power DENSITY
    
    powerspec[['freq','power']].query('freq < 267').to_csv('../data/bgfitting_PSDs/KIC'+star_name[4:]+'_PS.csv',
                                       index=False)
```

```python
highmass_rc_newvals = pd.read_csv('../datatables/highmass_rc_newvals.csv')
highmass_rc_newvals
```

```python
for i in range(0,len(highmass_rc_newvals)):
#     if highmass_rc['KIC'][i] in [10384595,3955502,6529078,2579142,10547065]:
    generate_emcee_files('KIC '+str(highmass_rc_newvals['KIC'][i]))
```

```python
for i in range(0,len(highmass_rc_newvals)):
#     if highmass_rc['KIC'][i] in [10384595,3955502,6529078,2579142,10547065]:
    generate_pysyd_files('KIC '+str(highmass_rc_newvals['KIC'][i]))
```

```python
pysyd_info = pd.read_csv('../pysyd-work/info/star_info.csv')
pysyd_info = pysyd_info.drop(pysyd_info.index[:])

for i in range(0,len(highmass_rc_newvals)):
    appending = pd.DataFrame({'star':highmass_rc_newvals['KIC'][i],#'star':'KIC'+str(highmass_rc['KIC'][i]),
                              'ind_width':1,
                              'lower_bg':1,
                              'numax':highmass_rc_newvals['numax'][i],
                              'smooth_ps':0.5,
#                               'verbose':True,
                              'lower_ex':1,
                              'smooth_width':5,
                              'binning':0.01}, index=[0])

    pysyd_info = pd.concat([pysyd_info, appending], ignore_index = True)
    
pysyd_info.to_csv('../pysyd-work/info/star_info.csv',index=False)
pysyd_info
```
