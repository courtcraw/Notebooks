```python
import numpy as np
import pandas as pd
```

```python
master = pd.read_csv('../rcb_master_list.csv')
```

```python
master_w_spec = master[master['Spectra_Type'].notnull()]
master_w_spec.reset_index(drop=True,inplace=True)
```

```python
def convert_radec_to_decimal(rahr,ramin,rasec,decdeg,decmin,decsec,suppress_output=False):
    #1 hr = 15 deg
    hours = rahr + (ramin/60) + (rasec/3600)
    ra_degrees = hours * 15
    if decdeg >= 0:
        dec_degrees = decdeg + (decmin/60) + (decsec/3600)
    else:
        dec_degrees = decdeg - (decmin/60) - (decsec/3600)
    if suppress_output == False:
        print('RA: %d hr %d min %.2f sec -> %.7f degrees' % (rahr,ramin,rasec,ra_degrees))
        print('Dec: %d : %d : %.2f  -> %.7f degrees' % (decdeg,decmin,decsec,dec_degrees))
    elif suppress_output == True:
        pass
    return ra_degrees, dec_degrees
```

```python
coords_list = []
for i in range(0,len(master_w_spec)-6):
    init_ra = master_w_spec['RA'][i]
    init_dec = master_w_spec['Dec'][i]
    
    ra1 = int(init_ra[0:2])
    ra2 = int(init_ra[3:5])
    ra3 = float(init_ra[6:])
    if init_dec[0] == '-':
        dec1 = -int(init_dec[1:3])
        dec2 = int(init_dec[4:6])
        dec3 = float(init_dec[7:])
    else:
        dec1 = int(init_dec[0:2])
        dec2 = int(init_dec[3:5])
        dec3 = float(init_dec[6:])
        
    store_ra,store_dec = convert_radec_to_decimal(ra1,ra2,ra3,dec1,dec2,dec3,suppress_output=True)
    coords_list.append([master_w_spec['Name'][i],np.round(store_ra,3),np.round(store_dec,3)])
    
coords_list
```

```python
coords_frame = pd.DataFrame(coords_list)
coords_frame
```

```python
coords_frame.to_csv('../hdc_coords')
```
