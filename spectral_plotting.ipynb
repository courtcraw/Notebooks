```python
def apply_extinction(lam, flux, extinction):
    ext = CCM89(Rv=3.1)
    wavelength_w_units = lam*u.Angstrom
    dereddened_flux = np.asarray(flux)/ext.extinguish(wavelength_w_units,Av=extinction)

    return lam, dereddened_flux
```

```python
blue_fits = pd.read_csv('../iras00483+6440_kastblue.dat',sep='\s+',names=['wavelength','flux','f_err'])
red_fits = pd.read_csv('../iras00483+6440_kastred.dat',sep='\s+',names=['wavelength','flux','f_err'])
blue_fits
```

```python
adopted_Av = 3.46
bluelam,blueflux = apply_extinction(np.asarray(blue_fits['wavelength']),np.asarray(blue_fits['flux']),
                                    adopted_Av)
plt.plot(bluelam,blueflux)
redlam,redflux = apply_extinction(np.asarray(red_fits['wavelength']),np.asarray(red_fits['flux']),
                                    adopted_Av)
plt.plot(redlam,redflux,color='tab:red')

plt.axvline(4922,color='k')
plt.axvline(4388,color='k')
plt.axvline(4471,color='k')
plt.axvline(3889,color='k')
plt.axvline(4026,color='k')
plt.axvline(4143,color='k')

plt.axvline(3968,color='g')
plt.axvline(3933,color='g')

# plt.plot(blue_fits['wavelength'],blue_fits['flux'],color='tab:blue')
# plt.plot(red_fits['wavelength'],red_fits['flux'],color='tab:red')
plt.xlim(3800,4500)
plt.savefig('J005128_carlmelis')
plt.show()
```
